# Multiple Linear Regression 
In this method, we have more than two variables as input that why it's called multiple linear regression.

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
# importing dataset 
df = pd.read_csv("ml_data_salary.csv")
df.head()
```

```python
# splitting data into X and y 
X = df[["age", "distance", "YearsExperience"]]
y = df["Salary"]
```

```python
# # splitting data into train and test variable for model training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 1/3, random_state=0)

```

```python
# # creating model and then training it
df = LinearRegression().fit(X_train, y_train)
df # our model is ready. Yahooo:)
```
